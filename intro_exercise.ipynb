```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('day.csv')
```

```python
# shows a preview of the data
df.head()
# shows some basic stats of the data
df.describe()
```

```python
# calculate the mean and standard deviation of the hourly data counts (the 'cnt' column)
df.cnt.mean()
```

```python
# standard deviation
df.cnt.std()
```

```python
# plot the counts ('cnt' column)
plt.plot(df.cnt)
plt.xlabel('Instance')
plt.ylabel('Hourly Count')
plt.title('Total Rented Bikes')
plt.show()
```
